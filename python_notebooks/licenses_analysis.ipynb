```python
# import packages
import pandas as pd
import altair as alt
```

```python
# import dataset
licenses = pd.read_csv("https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/licenses_fall2022.csv")
licenses.head()
```

```python
# counts of license types
license_types = licenses.groupby("License Type")["_id"].count()
license_types = license_types.reset_index(name="Count")
license_types = license_types.sort_values(by="Count", ascending=False)
license_types
```

```python
# bar chart of license types
chart1 = alt.Chart(license_types).mark_bar().encode(
    y="Count",
    x=alt.X("License Type").sort("y"),
    color=alt.Color("License Type", scale=alt.Scale(scheme="category10")),
    tooltip=["License Type", "Count"]
).interactive()

chart1

#chart1.properties(width="container").save("license_bar1.json")
```

```python
# license types
list(licenses["License Status"].unique())
```

```python
license_types_by_status = licenses.groupby(["License Status", "License Type"])["_id"].count()
license_types_by_status = license_types_by_status.reset_index(name="Count")
license_types_by_status = license_types_by_status.sort_values(by="Count", ascending=False)
license_types_by_status
```

```python
# status select dropdown
status_dropdown = alt.binding_select(options=license_types_by_status["License Status"].unique(), name="License Status: ")
status_select = alt.selection_point(fields=["License Status"], bind=status_dropdown, value="ACTIVE")

# bar chart of license types by status
chart2 = alt.Chart(license_types_by_status).mark_bar().encode(
    y="Count",
    x=alt.X("License Type").sort("y"),
    color=alt.Color("License Type", scale=alt.Scale(scheme="sinebow")),
    tooltip=["License Type", "Count"]
).add_params(status_select).transform_filter(status_select).interactive()

#chart2

chart2.properties(width="container").save("license_bar2.json")
```
